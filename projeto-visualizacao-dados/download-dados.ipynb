# Download de Dados

```python
import csv
```

```python
filename = "data/sitka_weather_07-2014.csv"

with open(filename) as f:
    reader = csv.reader(f)
    header_row = next(reader) # next devolve a próxima linha do arquivo quando recebe o objeto reader
    print(header_row)
```

```python
for index, column_header in enumerate(header_row): # usamos enumerate() na lista para obter o índice de cada item, assim como o valor
    print(index, column_header) 
```

Será extraido a temperatura máxima de cada dia e armazenadas de modo organizado em uma lista

```python
highs = []

with open(filename) as f:
    reader = csv.reader(f)
    header_row = next(reader) # next devolve a próxima linha do arquivo quando recebe o objeto reader
    for row in reader:
        high = int(row[1])
        highs.append(high)
        print(highs)
```

### Visualizando em um Gráfico

```python
from matplotlib import pyplot as plt
```

```python
fig = plt.figure(dpi=128, figsize=(10, 6))
plt.plot(highs, c='red')
plt.title("Daily high temperatures, July 2014", fontsize=24)
plt.xlabel('', fontsize=16)
plt.ylabel("Temperature (F)", fontsize=16) 
plt.tick_params(axis='both', which='major', labelsize=16)
plt.show()
```
